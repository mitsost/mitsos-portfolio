### 1. First things first: importing pandas, numpy, etc

```python
# Preliminary - importing modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

### 2. Reading the CSV file with the data
This csv file below (20k_wines.csv) contains a subset of the data from the original file 'cellartracker_edit.txt' -- which contains 2 million entries.
See "Final Project - Wine reviews - Dataframe creation" notebook on how this csv was created

```python
# Reading in the data from the CSV file
bigdf = pd.read_csv('20k_wines.csv')
```

```python
#Getting rid of empty data
bigdf = bigdf.dropna()
bigdf.count()
```

### 3. Cleaning the data

```python
# Removing 'N/A' values from the ratings column
print bigdf[bigdf['r.rating'] != ' N/A'].count()
bigdf2 = bigdf[bigdf['r.rating'] != ' N/A']
```

```python
print bigdf2[bigdf2["w.year"] != ' N/A'].count()
bigdf2 = bigdf2[bigdf2["w.year"] != ' N/A']
```

```python
bigfd2 = bigdf2.reset_index(drop=True)
bigdf2 = bigdf2.drop(bigdf2.columns[0], axis=1)
```

```python
bigdf2 = bigdf2.reset_index(drop=True)
```

```python
bigdf2
```

### 4. Plotting some variables to see if we discern any patterns

```python
x = bigdf2["w.year"]
y = bigdf2["r.rating"]
x = x.apply(lambda x: int(x))
y = y.apply(lambda x: int(x))
```

```python
plt.scatter(x, y)
plt.suptitle("Wine rating based on the vintage")
plt.xlabel("Vintage (year)")
plt.ylabel("Wine rating (0-100)")
```

```python
plt.hist(y)
plt.ylabel('Number of ratings')
plt.xlabel("Ratings")
plt.suptitle("Most wines were rated 90-95 points")
```

### Different wine types
Here we can see that most of the wines rated fall into the top ten, or so, types. Pinot Noir is the most commonly reviewed type

```python
dftypes = pd.DataFrame(bigdf2['w.type'].value_counts())
dftypes.head(20)
```

```python
df_pinot = bigdf2[bigdf2['w.type'] == " Pinot Noir"]
df_pinot
```

```python
y_pinot = df_pinot['r.rating']
y_pinot = y_pinot.apply(lambda x: int(x))
```

```python
plt.hist(y_pinot)
plt.ylabel('Number of ratings - Pinot Noir')
plt.xlabel("Ratings - Pinot Noir")
plt.suptitle("Ratings distribution for Pinot Noir")
```

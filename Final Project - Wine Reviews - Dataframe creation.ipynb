```python
# Preliminary - importing modules
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df1 = pd.read_csv('..\Downloads\cellartracker_edit_pipe.txt', sep='|', skiprows=0, nrows=9, header=None, error_bad_lines=False)
df1 = df1.transpose()
df1 = df1.drop(df1.index[0])
bigdf = df1
col_names = ['w.name', 'w.id', 'w.type', 'w.year', 'r.rating', 'r.time', 'r.userid', 'r.username', 'r.text']
bigdf.columns = col_names
bigdf
```

```python
for i in range(10, 200000, 10):
    #print i
    df3 = pd.read_csv('..\Downloads\cellartracker_edit_pipe.txt', sep='|', skiprows=i, nrows=9, header=None, engine='c', infer_datetime_format=True, error_bad_lines=False)
    #df3.columns = ['labels', 'data']
    #df3.set_index(['labels'])
    df3 = df3.transpose()
    df3 = df3.drop(df3.index[0])
    df3.columns = col_names
    bigdf = pd.concat([bigdf, df3.iloc[0:]], axis=0)
bigdf
```

```python
bigdf = bigdf.reset_index(drop=True)
```
